
### Q1. Find the equation of SVM using Lagrangian method and compute the equation of decision boundary.

The Lagrangian formulation for SVM is: `L(w,b,α) = (1/2)||w||² - Σᵢ αᵢ(yᵢ(w*xᵢ + b) - 1)`

For optimal solution, we take partial derivatives and set them to zero:
- ∂L/∂w = w - Σᵢ αᵢyᵢxᵢ = 0
- ∂L/∂b = -Σᵢ αᵢyᵢ = 0

The decision boundary equation will be:
#### f(x) = w*x + b = 0, where w = Σᵢ αᵢyᵢxᵢ

### Q2. Find out the hyper plane for the given set in the attached pic using SVM classifier.

Data Points: [1,1], [2,1], [1,-1], [2,-1], [4,0], [5,1], [5,-1], [6,0]

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Data points
X = np.array([[1,1], [2,1], [1,-1], [2,-1], [4,0], [5,1], [5,-1], [6,0]])
# Assuming points (1,1), (2,1), (5,1) are class 1 and others are class -1
y = np.array([1, 1, -1, -1, -1, 1, -1, -1])

# Training SVM
svm = SVC(kernel='linear')
svm.fit(X, y)

# Creating grid for plotting decision boundary
xx, yy = np.meshgrid(np.linspace(0, 7, 100),
                     np.linspace(-2, 2, 100))
Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot
plt.style.use('dark_background')
plt.figure(figsize=(10, 6))
plt.contour(xx, yy, Z, colors='w', levels=[-1, 0, 1], alpha=0.8,
           linestyles=['--', '-', '--'])
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.PiYG, s=150)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('SVM Decision Boundary')

# Plot support vectors
plt.scatter(svm.support_vectors_[:, 0], svm.support_vectors_[:, 1],
           s=150, linewidth=1, facecolors='none', edgecolors='w')
plt.grid(True)
```
